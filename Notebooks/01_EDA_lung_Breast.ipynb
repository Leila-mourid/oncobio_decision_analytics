### 9. Analyse par stade (Breast)

```python
sns.boxplot(x="6th Stage", y="Survival Months", data=breast)
plt.title("Survival by Stage - Breast")
plt.xticks(rotation=45)
plt.show()
```

### 10. Performance Status (Lung)

```python
sns.boxplot(x="ecog", y="survival_months", data=lung)
plt.title("Survival by ECOG - Lung")
plt.show()
```

### 11. Corrélations (Lung)

```python
import numpy as np

corr = lung.select_dtypes(include=np.number).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix - Lung")
plt.show()
```

### 12. Comparaison multi-cancers

```python
breast_temp = breast.copy()
breast_temp["cancer_type"] = "Breast"
breast_temp = breast_temp.rename(columns={"Survival Months":"survival_months"})

lung_temp = lung.copy()
lung_temp["cancer_type"] = "Lung"

common = pd.concat([
    breast_temp[["Age","survival_months","cancer_type"]].rename(columns={"Age":"age"}),
    lung_temp[["age","survival_months","cancer_type"]]
])

sns.boxplot(x="cancer_type", y="survival_months", data=common)
plt.title("Survival Comparison Breast vs Lung")
plt.show()
```
